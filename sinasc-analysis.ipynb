### Exploratory analysis

#### In this notebook, I analyze the microdata for all live births in which the mother was **14 years old or younger** between 2004 and 2024.

#### The goal is to calculate the main descriptive statistics used throughout the project, including the number of births over time, rates per 1,000 live births, geographic distribution, and the demographic profile of the girls (such as age, race/color, and marital status).

```python
import pandas as pd

df = pd.read_csv("microdados_sinasc_2004_2024.csv")
df.head()
```

```python
df.shape
```

```python
df.sort_values("ano", inplace=True)
df
```

### Number of cases per year (for the line chart)

```python
cases_per_year = df.groupby("ano").size()
cases_per_year
```

### Number of cases per brazilian state

```python
cases_per_state = (
    df.groupby("sigla_uf")
      .size()
      .reset_index(name="total_casos")
      .sort_values(by="total_casos", ascending=False)
)

cases_per_state
```

```python
cases_per_state = (
    df[df["ano"] == 2024]
      .groupby("sigla_uf")
      .size()
      .reset_index(name="total_casos")
      .sort_values(by="total_casos", ascending=False)
)

cases_per_state
```

### Number of cases per city

```python
cases_per_city = df.groupby("id_municipio_nascimento_nome").size()
cases_per_city.sort_values(ascending=False)
```
